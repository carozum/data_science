```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

path = os.path.join("data", "taylor_swift_spotify.csv")
df = pd.read_csv(path, index_col=0)
```

## Formattage de la release date into a datetime

```python
# changement de la release_date en datetime
df['release_date'] =pd.to_datetime(df['release_date'])
```

```python
# création d'une colonne Year et Month
df['year'] = df['release_date'].dt.year 
df['month'] = df['release_date'].dt.month
df.head(2)
```

```python
df.info()
```

## ALBUMS

```python
# Nettoyage de la colonne album
df['album'] = [album.replace(" (Taylor's Version)", '') for album in df['album']]
df['album'] = [album.replace(" (The Til Dawn Edition)", '') for album in df['album']]
df['album'] = [album.replace(" (International Version)", '') for album in df['album']]
df['album'] = [album.replace(" (3am Edition)", '') for album in df['album']]
df['album'] = [album.replace(" (deluxe version)", '') for album in df['album']]
df['album'] = [album.replace(": the long pond studio sessions (from the Disney+ special) [deluxe edition]", '') for album in df['album']]
df['album'] = [album.replace(" Stadium Tour Surprise Song Playlist", '') for album in df['album']]
df['album'] = [album.replace(" (deluxe version)", '') for album in df['album']]
df['album'] = [album.replace(" (Deluxe Package)", '') for album in df['album']]
df['album'] = [album.replace(" (Deluxe Edition)", '') for album in df['album']]
df['album'] = [album.replace(" World Tour Live", '') for album in df['album']]
df['album'] = [album.replace(" (Platinum Edition)", '') for album in df['album']]
df['album'] = [album.replace(" (Deluxe)", '') for album in df['album']]
df['album'] = [album.replace("Live From Clear Channel Stripped 2008", 'Live 2008') for album in df['album']]
#df
```

```python
albums = list(df['album'].unique())
```

```python
pivot = df.pivot_table(index=['year'], values ='popularity', aggfunc='mean')
pivot
```
